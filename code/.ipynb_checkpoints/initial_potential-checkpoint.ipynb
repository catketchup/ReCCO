```python
# 3-d k-space axes
k_x = np.arange(-r_o/d, (r_o+d)/d)/(d*N)
k_y = np.arange(-r_o/d, (r_o+d)/d)/(d*N)
k_z = np.arange(-r_o/d, (r_o+d)/d)/(d*N)
# 3-d meshgrid k-space coordinate
k_xx, k_yy, k_zz = np.meshgrid(k_x, k_y, k_z)
# as a comparison with np.fft.fftfreq(21, 10)
kk = np.sqrt(k_xx**2 + k_yy**2 + k_zz**2)
```

```python
kk_flat = kk.flatten()
xi_k_flat = xi_k.flatten()
Ppsi_k = np.reshape(F_k_itp(kk_flat)*xi_k_flat, kk.shape)
# Ppsi_k = np.reshape(xi_k_flat, kk.shape)
Ppsi = np.fft.ifft(Ppsi_k)
```

```python
# check P_psi_k https://bertvandenbroucke.netlify.app/2019/05/24/computing-a-power-spectrum-in-python/
kbins = np.linspace(1e-10, 1e-2, 100)

import scipy.stats as stats
Ppsi_k_amplitude = np.abs(Ppsi_k)**2
Abins, _, _ = stats.binned_statistic(kk.flatten(), Ppsi_k_amplitude.flatten(),
                                     statistic = "mean",
                                     bins = kbins)
kvals = 0.5 * (kbins[1:] + kbins[:-1])
Abins *= np.pi * (kbins[1:]**2 - kbins[:-1]**2)
plt.plot(kvals, np.sqrt(Abins))
plt.xscale('log')
# plt.plot(k_test, P_psi_k)
```
